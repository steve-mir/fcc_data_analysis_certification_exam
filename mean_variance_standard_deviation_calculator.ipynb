```python
import numpy as np
import urllib.request
```

```python

# Mean
def get_mean(list):
    a1 = []
    a2 = []
    fl = 0
    arr = np.array(list)
    mainArr = np.array([[], [], 0])
    m3x3 = arr.reshape(3, 3)
    
    i = 0
    while i < 3:
        # mean
        a1.append(m3x3[:, i].mean())
        a2.append(m3x3[i].mean())

        fl = arr.mean()
        mainArr[0] = a1
        mainArr[1] = a2
        mainArr[2] = fl
        i += 1
    
    return mainArr.tolist()


def get_var(list):
    a1 = []
    a2 = []
    fl = 0
    arr = np.array(list)
    mainArr = np.array([[], [], 0])
    m3x3 = arr.reshape(3, 3)

    i = 0
    while i < 3:
        # variance
        a1.append(m3x3[:, i].var())
        a2.append(m3x3[i].var())
        
        fl = arr.var()
        mainArr[0] = a1
        mainArr[1] = a2
        mainArr[2] = fl
        i += 1
    return mainArr.tolist()
    


def get_std(list):
    a1 = []
    a2 = []
    fl = 0
    arr = np.array(list)
    mainArr = np.array([[], [], 0])
    m3x3 = arr.reshape(3, 3)

    i = 0
    while i < 3:
        # standard deviation
        a1.append(m3x3[:, i].std())
        a2.append(m3x3[i].std())
            
        fl = arr.std()
        mainArr[0] = a1
        mainArr[1] = a2
        mainArr[2] = fl
        i += 1

    return mainArr.tolist()


def get_max(list):
    a1 = []
    a2 = []
    fl = 0
    arr = np.array(list)
    mainArr = np.array([[], [], 0])
    m3x3 = arr.reshape(3, 3)

    i = 0
    while i < 3:
        # sum
        a1.append(m3x3[:, i].max())
        a2.append(m3x3[i].max())
            
        fl = arr.max()
        mainArr[0] = a1
        mainArr[1] = a2
        mainArr[2] = fl
        i += 1
        
    return mainArr.tolist()


def get_min(list):
    a1 = []
    a2 = []
    fl = 0
    arr = np.array(list)
    mainArr = np.array([[], [], 0])
    m3x3 = arr.reshape(3, 3)

    i = 0
    while i < 3:
        # sum
        a1.append(m3x3[:, i].min())
        a2.append(m3x3[i].min())
            
        fl = arr.min()
        mainArr[0] = a1
        mainArr[1] = a2
        mainArr[2] = fl
        i += 1
        
    return mainArr.tolist()


def get_sum(list):
    a1 = []
    a2 = []
    fl = 0
    arr = np.array(list)
    mainArr = np.array([[], [], 0])
    m3x3 = arr.reshape(3, 3)

    i = 0
    while i < 3:
        # sum
        a1.append(m3x3[:, i].sum())
        a2.append(m3x3[i].sum())
            
        fl = arr.sum()
        mainArr[0] = a1
        mainArr[1] = a2
        mainArr[2] = fl
        i += 1
    return mainArr.tolist()
    
```

```python
# axis 1[:,i] = column
# axis 2[i] = row
# flattened = list
def calculate(list):
    cal = {}
    cal['mean'] = get_mean(list)
    cal['variance'] = get_var(list)
    cal['standard deviation'] = get_std(list)
    cal['max'] = get_max(list)
    cal['min'] = get_min(list)
    cal['sum'] = get_sum(list)
    
    return cal
```

```python
calculate([0,1,2,3,4,5,6,7,8])
```
